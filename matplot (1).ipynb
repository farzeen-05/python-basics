```python
#styling line graphs
a=[1,2,3,4,5]
b=[7,8,9,6,7]
plt.plot(a,b,color='violet',marker='d',markersize='8',markeredgecolor='g',markerfacecolor='b',linestyle=':')
plt.show()
#red-r,black-b,yellow-y,green-g
#o-circle,s-square,D,d-diamond,H,h-hexagon,><^-'Triangle',+*-star
#line style--:,-,--,-.
```

```python
#short method - line color,marker style,line style
a=[1,2,3,4,5]
b=[45,21,65,11,89]
plt.plot(a,b,"bo:",markeredgecolor='y')
plt.show()
```

```python
#title,x label,y label,fig size
a=[66,77,88,99,70,44]
b=['s1','s2','s3','s4','s5','s6']
plt.figure(figsize=(10,5),dpi=100)
plt.plot(b,a,'bD:')
plt.title("patient age data",color='r')
plt.xlabel("patient ID",color='purple')
plt.ylabel("patient age",color='y')
plt.grid(color='y')
plt.show()
```

```python
#fontdict,saving graph
student_id=['s1','s2','s3','s4','s5']
screen_time=[1,2,3,4,5]
plt.plot(student_id,screen_time)
plt.title("student screen time data",fontdict={'family':'serif','color':'purple',
                                               'weight':'bold','size':30})
plt.xlabel("student ID",fontdict={'family':'serif','color':'y',
                                               'weight':'bold'})          
plt.ylabel("screen time",fontdict={'family':'serif','color':'y',
                                               'weight':'bold'})
plt.savefig('graph.png',dpi=100)
plt.show()
```

```python
#time Series data-for continues data-line graph,histogram,scatter plot
#categorical data-bar chart,pie chart,horizontal
```

```python
#bar chart
subjects=['c','c++','java','python','html']
marks=[55,66,66,77,88]
c=['b','r','y','black','purple']
plt.bar(subjects,marks,color=c)
plt.show()
```

```python
#horizontal chart
subjects=['c','c++','java','python','html']
marks=[55,66,66,77,88]
c=['b','r','y','black','purple']
plt.barh(subjects,marks,color=c)
plt.show()
```

```python
#scatter plot
age=[33,18,19,20,21]
patient_id=['p1','p2','p3','p4','p5']
c=['b','r','y','black','purple']
plt.scatter(patient_id,age,color=c)
plt.show()
```

```python
#histogram
age=[22,34,55,66,66,77,74,22,55,66,77,77,99,100,222,233]
a=[0,50,100,250]
plt.hist(age,bins=a,color='r')
plt.show()
```

```python
#pie chart
marks=[33,18,19,20,21]
a=[0.4,0.1,0.2,0.3,0.35]
person=['farzeen','mugni','shahim','nasir','moosa']
plt.pie(marks,labels=person,autopct='%1.2f%%',explode=a)
a=['farzeen','mugni','shahim','nasir','moosa']
plt.legend(a,loc='upper left')
plt.show()
```

```python
#advertisement effect on sales
x1=[33,44,55,66,77]
y1=[11,22,33,44,55]
x2=[44,22,33,11,22]
y2=[22,44,55,66,77]
plt.plot(x1,y1,'gH-.')
plt.plot(x2,y2,'bH-.')
plt.title("advertisement effect on sales",fontdict={'family':'serif','color':'purple',
                                               'weight':'bold','size':20})
plt.xlabel("medium")
plt.ylabel("sales")
a=['tv','mobiles']
plt.legend(a,loc='center left')
plt.grid()
plt.show()
```
