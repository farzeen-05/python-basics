```python
#task 1
import numpy as np
data = { 
"City": ["Delhi", "Mumbai", np.nan, "Chennai"], 
"Popula on": [19, 20, 15, np.nan], 
"Area": [1484, np.nan, 426, 1189] 
} 
df=pd.DataFrame(data)
print(df)
print(df.isna().sum())
print(df.dropna(axis=0))
print(df.dropna(axis=1))
print(df.dropna(axis=1,inplace=True))
print(df)
```

```python
#task 2
data = { 
"City": ["Delhi", "Mumbai", np.nan, "Chennai"], 
"Population": [19, 20, 15, np.nan], 
"Area": [1484, np.nan, 426, 1189] 
} 
df=pd.DataFrame(data)
print(df.fillna(0))
print(df.ffill())
print(df.bfill())
b=df["Population"].mean()
print(df["Population"].fillna(b,inplace=True))
c=df["Area"].median()
print(df["Area"].fillna(c,inplace=True))
print(df)
```
