```python
#one hot encoding
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop=None,sparse=False) #sparse output 
encoded = encoder.fit_transform(df[['gender']]) #apply encoding to our column

#create data frame with new columns
encoded_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out(['gender']))
print(encoded_df) 

```

```python
#concatenate with original data frame (dropping old columns)
df=pd.concat([df.drop('gender',axis=1),encoded_df],axis=1)
df.head()
```

```python
#label encoding
#step 1-import 
from sklearn.preprocessing import LabelEncoder

#step 2- create object for class
lb=LabelEncoder()

#step 3- pass the data to transform
df['smoking_status']=lb.fit_transform(df['smoking_status'])
df
```

```python
#mapping-map()-original data (can be used it for nominal data)
df['gender'].unique()
df['gender']=df['gender'].map({'female':0,'male':1,'other':2})
```
