```python
import numpy as np
#indexing and slicing in arrays 

#1D
a=np.array([5,6,7,8,9,7])

#indexing
print(a[1])
print(a[-4])

#slicing
print(a[:])
print(a[::2])
print(a[::-2])
print(a[2:5:3])
```

```python
#2D arrays
a=np.array([[2,3,4],[3,3,5],[3,5,6]])
print(a)

#printing single elements
print(a[2,1])
print(a[0,0])
print(a[-1,-1])

#print rows
print(a[1])
print(a[-1])
print(a[0:3:2])

#print columns
print(a[:,0])
print(a[:,2])
print(a[:2,2])
print(a[::2,2])
print(a[:,:])

#print submatrix
print(a[0:2,0:2])
print(a[1:2,0:2])
```

```python
#copy and view

#copy
a=np.array([2,3,4,5,6])
b=a.copy()
print(a)
print(b)
b[1]=101
print(b)
print(a)
```

```python
#view
a=np.array([2,3,4,5,6])
b=a.view()
print(a)
print(b)
b[3]=400
print(b)
print(a)
```

```python
#reshaping is method of converting one dimensional to other dimensional
x=np.arange(1,13)
print(x)
print("dim",x.ndim)

y=x.reshape(3,4)
print(y)
print("dim",y.ndim)
```

```python
x=np.arange(1,121)
print(x)

y=x.reshape(4,-1)
print(y)

z=x.reshape(-1,10)
print(z)
```

```python
#reshaping 2D to 1D
#flatten()- creates copy of original values
#ravel()- creates view of original values
a=np.array([[2,3,4],[6,7,8],[4,5,6],[8,9,8]])
print(a)
print("dim",a.ndim)
#flatten
b=a.flatten()
print(b)
print("dim",b.ndim)
b[1]=100
print(b)
print(a)
```

```python
#ravel
c=a.ravel()
print(c)
print("dim:",c.ndim)

c[1]=100
print(c)
print(a)
```

```python
#task-create 2 3D arrays and use flatten() and ravel() method for both and check dimensions
import numpy as np  
a=np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
b=np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print("adim",a.ndim)
print("bdim",b.ndim )
c=a.flatten()
print(c)
d=b.flatten()
print(d)
print("dim",c.ndim)
print("dim",d.ndim)
e=a.ravel()
f=b.ravel()
print(e)
print(f)
print("dim",e.ndim)
print("dim",f.ndim)
```
