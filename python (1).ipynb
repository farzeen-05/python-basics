```python
#task 1
#a.t.m implement using function and class
class ATM:
    def __init__(self, correct_pin, balance):
        self.correct_pin = correct_pin
        self.balance = balance
    def login(self, pin):
        if pin == self.correct_pin:
            print("Login Successful")
            return True
        else:
            print("Incorrect PIN")
            print("Program Closed")
            return False
    def credit(self, amount):
        if amount > 0:
            self.balance += amount
            print("Amount credited successfully")
            print("Updated balance:", self.balance)
        else:
            print("Invalid amount")
    def debit(self, amount):
        if amount > 0:
            if self.balance - amount >= 0:
                self.balance -= amount
                print("Amount debited successfully")
                print("Updated balance:", self.balance)
            else:
                print("Insufficient balance")
        else:
            print("Invalid amount")
    def balance_check(self):
        print("Current balance:", self.balance)
    def run(self):
        while True:
            print("\n1. Credit")
            print("2. Debit")
            print("3. Balance Check")
            print("4. Exit")
            choice = int(input("Enter your choice: "))
            if choice == 1:
                amount = int(input("Enter amount to credit: "))
                self.credit(amount)
            elif choice == 2:
                amount = int(input("Enter amount to debit: "))
                self.debit(amount)
            elif choice == 3:
                self.balance_check()
            elif choice == 4:
                print("Thank you, program closed")
                break
            else:
                print("Invalid choice")
atm = ATM(correct_pin=1234, balance=10000)
pin = int(input("Enter the PIN: "))
if atm.login(pin):
    atm.run()
```

```python
class StudentManagement:
    def __init__(self):
        self.student={}
    def add_student(self):
        name=input("enter the name of students:")
        marks=int(input("eneter the marks of student:"))
        self.student[name]=marks
        print(f"{name} added with {marks}marks")
    def view_student(self):
        if not self.student:
            print("no student found")
        else:
            print("student list")
            for name,marks in self.student.items():
                print(f"{name}-{marks}")
    def new_marks(self):
        name=input("enter the name of students:")
        if name in self.student:
            new_marks=int(input("enter the marks:"))
            self.student[name]=new_marks
            print(f" {name} with {new_marks}")
def main():
    system=StudentManagement()
    print("welcome to student management system:")
    while True:
        print("\n menu ")
        print("1.add students")
        print("2.view all students")
        print("3.update marks")
        print("4.exit")
        choice=int(input("enter the choice:"))
        if choice==1:
            system.add_student()
        elif choice==2:
            system.view_student()
        elif choice==3:
            system.new_marks()
        elif choice==4:
            print("thank you closing the program")
            break
        else:
            print("invalid choice")

main()
    
```
